```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfs = {}

dfs['AEE'] = pd.read_csv('AEE.csv')
dfs['FRT'] = pd.read_csv('FRT.csv')
dfs['HD'] = pd.read_csv('HD.csv')
dfs['MTD'] = pd.read_csv('MTD.csv')
dfs['UNH'] = pd.read_csv('UNH.csv')
```

```python
dfs['AEE']
```

```python
for name, df in dfs.items():
    df["Date"] = pd.to_datetime(df["Date"])
    
```

```python
#1 
for name, df in dfs.items():
    print(name, "sum of rows with identical high and low:",(df["High"] == df["Low"]).sum())
```

There is no company stock that would have identical high and low on any day. It would mean that in that day the stock price didn't move at all.

```python
#2

for name, df in dfs.items():
    print("Maximum price recorded for",name,"is",df['High'].max())
    print("Minimum price recorded for",name,"is",df['Low'].min())
    print()
    
```

```python
#3
for name, df in dfs.items():
    df["log_return"] = np.log(df["Close"] / df["Close"].shift(1))
    
```

```python
dfs['UNH']
```

```python
for name, df in dfs.items():
    print("Min log returns for",name,"is",df["log_return"].min())
    print("Max log returns for",name,"is",df["log_return"].max())
    print("Mean of log returns for",name,"is",df["log_return"].mean())
    print("Standard deviation of log return for",name,"is",df["log_return"].std())
    print()
```

```python
for name, df in dfs.items():
    df.set_index("Date", inplace=True)
```

```python
dfs
```

```python
#4

for name, df in dfs.items():
    print("Second highest return for",name,"was at",df["log_return"].nlargest(2).index)
    print("Second highest return for",name,"was at",df["log_return"].nsmallest(2).index)
    print()
```

```python
#5
for name, df in dfs.items():
 
    weekly_avg = df["Volume"].resample("W").mean().mean()
    
    print(name, weekly_avg)
```

```python
#6
```

```python
#7
for name, df in dfs.items():
    df['Volume Class'] = pd.qcut(df['Volume'], 3, labels=False)
```

```python
dfs
```

```python
#8

for name, df in dfs.items():
    plt.plot(df.index, df["log_return"], label=name, alpha = 0.4)

plt.legend()
plt.title("Log Returns of Companies")
plt.xlabel("Date")
plt.ylabel("Log Return")
plt.show()
```
